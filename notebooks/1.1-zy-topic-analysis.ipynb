```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
daily_topics = pd.read_csv('../data/processed/daily_topics.csv')
influencer_daily_topics = pd.read_csv('../data/processed/influencer_daily_topics.csv')
normal_daily_topics = pd.read_csv('../data/processed/normal_daily_topics.csv')
```

```python
def get_topics_df(df):
    all_topics = df.topic_0 + df.topic_1 + df.topic_2
    all = ''.join(list(all_topics))
    all_topic_list = re.findall(r"'(.*?)'", all)
    return pd.DataFrame(all_topic_list)
```

```python
daily_topic_df = get_topics_df(daily_topics)
inf_daily_topic_df = get_topics_df(influencer_daily_topics)
normal_daily_topic_df = get_topics_df(normal_daily_topics)
```

# Visualization

```python
def generate_wordcloud(df):
    wordcloud = WordCloud(background_color='white',
                          max_words=200,
                          max_font_size=40,
                          scale=3,
                          random_state=1,
                          colormap='viridis').generate(' '.join(list(df[0])))
    plt.figure(figsize=(20, 10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
generate_wordcloud(daily_topic_df)
```

```python
generate_wordcloud(inf_daily_topic_df)
```

```python
generate_wordcloud(normal_daily_topic_df)
```
